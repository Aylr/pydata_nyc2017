3. What happens if $\alpha$ is huge for Ridge or Lasso?
4. What happens if $\alpha$ is 0 for Ridge or Lasso?

$$C(\mathbf{w}) = \sum_j (\mathbf{x}_j \cdot \mathbf{w} - y_j)^2 + \alpha \sum_j w_j^2$$

$$C(\mathbf{w}) = \sum_j (\mathbf{x}_j \cdot \mathbf{w} - y_j)^2 + \alpha \sum_j \left|w_j\right|$$

```python
from sklearn.linear_model import Ridge, Lasso

import utils

x_train, x_test, y_train, y_test, X_train, X_test, transform, scale = utils.get_overfitting_data()
```

```python
ridge_reg = Ridge(alpha=10000).fit(X_train, y_train)
utils.plotter(x_test, y_test, fit_fn=ridge_reg.predict, transform=transform)
```

```python
ridge_reg = Ridge(alpha=0).fit(X_train, y_train)
utils.plotter(x_test, y_test, fit_fn=ridge_reg.predict, transform=transform)
```

```python
lasso_reg = Lasso(alpha=10000).fit(X_train, y_train)
utils.plotter(x_test, y_test, fit_fn=lasso_reg.predict, transform=transform)
```

```python
lasso_reg = Lasso(alpha=0.0000001, max_iter=10000000).fit(X_train, y_train)
utils.plotter(x_test, y_test, fit_fn=lasso_reg.predict, transform=transform)
```
